```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("laptop_price.csv", encoding="latin-1")
```

```python
df
```

```python
df.Company.value_counts
```

```python
df.drop("Product", axis = 1, inplace = True)
```

```python
df = df.join(pd.get_dummies(df.Company))
df = df.drop("Company",axis = 1)
```

```python
df = df.join(pd.get_dummies(df.TypeName))
df = df.drop("TypeName", axis = 1)
```

```python
df
```

```python
split_res = df.ScreenResolution.str.split()
split_res
```

```python
def extract_res(resolution_list):
    for val in resolution_list:
        if 'x' in val:
            return val
    return None
```

```python
extracted_res = split_res.apply(extract_res)
extracted_res
```

```python
df['width'] = extracted_res.str.split('x').str[0]
df['height'] = extracted_res.str.split('x').str[1]
```

```python
df
```

```python
df.drop("ScreenResolution",axis=1, inplace = True)
```

```python
df['CPU Brand'] = df.Cpu.str.split(" ").apply(lambda x: x[0])
df['CPU Frequency'] = df.Cpu.str.split(" ").str[-1]
```

```python
df.drop("Cpu", axis = 1, inplace = True)
```

```python
df['CPU Frequency'] = df["CPU Frequency"].str[:-3]
df
```

```python
df["Ram"] = df["Ram"].str[:-2]
```

```python
df["Ram"] = df["Ram"].astype("int")
df["CPU Frequency"] = df["CPU Frequency"].astype("float")
```

```python
df["width"] = df["width"].astype("int")
df["height"] = df["height"].astype("int")
```

```python
# df.hist()
```

```python
# df.dtypes
```

```python
bool_cols = df.select_dtypes(include = ['bool']).columns
bool_cols
```

```python
for col in bool_cols:
    df[col] = df[col].astype(int)
```

```python
# df.hist()
```

```python
df['Memory Amount'] = df.Memory.str.split(" ").apply(lambda x: x[0])
df["Memory type"] = df.Memory.str.split(" ").apply(lambda x: x[1])
```

```python
# df["Memory type"].value_counts
# df["Memory Amount"].value_counts
```

```python
def turn_memory_into_mb(value):
    if "GB" in value:
        return float(value[:value.find("GB")]) * 1000
    elif "TB" in value:
        return float(value[:value.find("TB")]) * 1000000
```

```python
df["Memory Amount"] = df["Memory Amount"].apply(turn_memory_into_mb)
```

```python
df.drop("Memory", axis = 1, inplace = True)
```

```python
df
```

```python
df["Weight"] = df["Weight"].str[:-2]
```

```python
df["Weight"] = df["Weight"].astype("float")
```

```python
df["GPU Brand"] = df.Gpu.str.split(" ").apply(lambda x: x[0])
```

```python
df = df.drop("Gpu", axis = 1)
```

```python
df = df.join(pd.get_dummies(df.OpSys))
df = df.drop("OpSys", axis = 1)
```

```python
df
```

```python
# you have also created same function above
boolcols = df.select_dtypes(include = ['bool']).columns
boolcols
```

```python
for col in boolcols:
    df[col] = df[col].astype(int)
```

```python
cpu_categories = pd.get_dummies(df["CPU Brand"])
cpu_categories.columns = [col + "_CPU" for col in cpu_categories.columns]

df = df.join(cpu_categories)
df = df.drop("CPU Brand", axis = 1)
```

```python
gpu_categories = pd.get_dummies(df["GPU Brand"])
gpu_categories.columns = [col + "_GPU" for col in gpu_categories.columns]

df = df.join(gpu_categories)
df = df.drop("GPU Brand", axis = 1)
```

```python
boolcol = df.select_dtypes(include = ['bool']).columns
boolcol
```

```python
for col in boolcol:
    df[col] = df[col].astype(int)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor

X, y = limited_df.drop("Price_euros", axis = 1), limited_df["Price_euros"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.15)

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['auto', 'sqrt', 'log2']
}
```

```python
forest = RandomForestRegressor(random_state=42)
tree = DecisionTreeRegressor()
knn = KNeighborsRegressor()

tree.fit(X_train_scaled, y_train)
knn.fit(X_train_scaled, y_train)
```

```python
from sklearn.model_selection import GridSearchCV
grid_search = GridSearchCV(estimator=forest, 
                           param_grid=param_grid, 
                           cv=5, 
                           scoring='r2',
                           n_jobs=-1) 
```

```python
grid_search.fit(X_train_scaled, y_train)
```

```python
best_rf = grid_search.best_estimator_
y_pred = best_rf.predict(X_test_scaled)
y_pred1 = tree.predict(X_test_scaled)
y_pred2 = knn.predict(X_test_scaled)
```

```python
# print(forest.score(X_test_scaled, y_test))
# print(tree.score(X_test_scaled, y_test))
# print(knn.score(X_test_scaled, y_test))
```

```python
from sklearn.metrics import r2_score, mean_absolute_error
print(r2_score(y_test, y_pred))
print(r2_score(y_test, y_pred1))
print(r2_score(y_test, y_pred2))
```

```python
print(mean_absolute_error(y_pred, y_test))
print(mean_absolute_error(y_pred1, y_test))
print(mean_absolute_error(y_pred2, y_test))
```

```python
plt.figure(figsize=(12,8))
plt.scatter(y_pred, y_test, marker = 'o', label = 'Random Forest')
plt.scatter(y_pred1, y_test, marker = 'D', label = 'Decision Tree')
plt.scatter(y_pred2, y_test, marker = '+', label = 'KNN')

plt.plot(range(0,6000), range(0,6000),c = "red")
plt.xlabel("Predicted values")
plt.ylabel("Actual values")
plt.title("Comparison of predicted vs actual values")

plt.legend()
```
